```python
import pandas as pd 
import numpy as np
import io
from nltk.corpus import stopwords 
from nltk.tokenize import word_tokenize
stop_words = set(stopwords.words('english'))
import string

#Open file
jeopardy_df = pd.read_csv('JEOPARDY_CSV.CSV')
jeopardy_df.head()

#Lower case
jeopardy_df.columns = jeopardy_df.columns.str.replace(' ', '')
jeopardy_df.Question = jeopardy_df.Question.str.lower()
#jeopardy_df.Answer = jeopardy_df.Answer.str.lower() #No need to lower as a lot of answers are names

#Tokenise
jeopardy_df['tokenised'] = jeopardy_df['Question'].apply(word_tokenize)

#Remove punctuation
jeopardy_df['nopunct'] = jeopardy_df['tokenised'].apply(lambda x: [char for char in x if char not in string.punctuation])

#Remove stop words
jeopardy_df['nostop'] = jeopardy_df['nopunct'].apply(lambda x: [item for item in x if item not in stop_words])

#Lemmatise
lemmatizer = nltk.stem.WordNetLemmatizer()
jeopardy_df['lemma'] = jeopardy_df['nostop'].apply(lambda x: [lemmatizer.lemmatize(w) for w in x])

jeopardy_df.head(20)
```

```python
from sklearn.feature_extraction.text import CountVectorizer

df = jeopardy_df[['lemma', 'Answer']].copy()
print(df.shape)

bow = CountVectorizer()
transform = bow.fit_transform(df)
print(transform.shape)
print(bow.get_feature_names)
pd.DataFrame(transform.toarray())
```
