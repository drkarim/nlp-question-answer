```python
import io 
from nltk.corpus import stopwords 
from nltk.tokenize import word_tokenize 
nltk.download('popular')
```

```python
jeopardy_df.columns = jeopardy_df.columns.str.replace(' ', '')
```

```python
stop_words = set(stopwords.words('english')) 
```

```python
jeopardy_df2 = jeopardy_df.copy()
```

```python
jeopardy_df2.head() 
```

```python
list(jeopardy_df)
```

```python
jeopardy_df2['Question'].apply(lambda x: [item for item in x if item not in stop_words])
```

```python
jeopardy_df2['Question_nostop'] = jeopardy_df2['Question'].apply(lambda x: ' '.join([word for word in x.split() if word not in (stop_words)]))
```

```python
jeopardy_df2.head() 
```

```python
w_tokenizer = nltk.tokenize.WhitespaceTokenizer()
lemmatizer = nltk.stem.WordNetLemmatizer()
from nltk.stem.snowball import SnowballStemmer
stemmer = SnowballStemmer("english")

def lemmatize_text(text):
    return [lemmatizer.lemmatize(w) for w in w_tokenizer.tokenize(text)]
```

```python
jeopardy_df2['Question_lemma'] = jeopardy_df2['Question_nostop'].apply(lambda x: ' '.join(lemmatize_text(x)))
```
